```python
#renaming the column from movie to title.
budgets = budgets.rename(columns={"movie": "title"})
```

```python
# checking for data types in our table.
movies2.info()
```

```python
#Converting production budget column into float datatype.
movies2["production_budget"] = movies2["production_budget"].apply(
    lambda x: 0 if x == 0 else float(x.split()[0].replace(",", "").replace("$", "")))  
# converting domestic gross column into float datatype
movies2["domestic_gross"] = movies2["domestic_gross"].apply(
    lambda x: 0 if x == 0 else float(x.split()[0].replace(",", "").replace("$", "")))
  # converting worldwide gross column into float datatype
movies2["worldwide_gross"] = movies2["worldwide_gross"].apply(
    lambda x: 0 if x == 0 else float(x.split()[0].replace(",", "").replace("$", "")))
```

### Finding the average rating per genre for the sample data above.

```python
#joining two tables movie_basics and movie_ratings to show average rating of Action Movies
table1 = """
SELECT primary_title, genres, AVG(averagerating) AS Action_average
FROM movie_basics
JOIN movie_ratings
    ON movie_basics.movie_id = movie_ratings.movie_id
    WHERE genres='Action'
    ORDER BY averagerating;
"""
data = pd.read_sql(table1, conn)
plt.bar(data.genres, data.Action_average)
plt.ylabel("Action Average")
plt.title("ACTION genre")
plt.show()
```

### This Visualization shows the Mean of average ratings of Action Genre of Movies AS 5.8

```python
#joining two tables movie_basics and movie_ratings to show average rating of Drama Movies
table2 = """
SELECT primary_title, genres, AVG(averagerating) AS Drama_Average
FROM movie_basics
JOIN movie_ratings
    ON movie_basics.movie_id = movie_ratings.movie_id
    WHERE genres='Drama'
    ORDER BY averagerating;
"""
data = pd.read_sql(table2, conn)
plt.bar(data.genres, data.Drama_Average)
plt.ylabel("Drama Average")
plt.title("Drama genre")
plt.show()
```

```python
#reading the movie_basics table.
q5 = pd.read_sql(
    """
SELECT genres, runtime_minutes, primary_title AS title
FROM movie_basics
""",
    conn,
)
```

### Joining the two tables 

```python
# Join the Number  data with IMDB data and take only those where both records exist
mainmovies = movie_gross.merge(q5, how="inner", on="title")
mainmovies.sort_values(by=("total_gross"), ascending=False).head(10) 
#checking for duplicates in 
mainmovies = movie_gross.drop_duplicates(subset="title")
 # converting total gross into millions
mainmovies["total_gross"] = (mainmovies["total_gross"] / 1000000)
 # Looking at movies per year.
years = (mainmovies.groupby(by="year")["title"].count().reset_index().head(10))
# Sort movies by studio and show number of movies per studio and the sum of total gross.
studios = (mainmovies.groupby(["studio"]).agg({"title": ["count"], "total_gross": ["sum"]}).reset_index())
studios.columns = ["Studio", "Movie_Count", "Total_Gross"]
studios = studios.sort_values(by=["Total_Gross"], ascending=False)
studioss = studios.head(10)
```

```python
# plot count and gross per studio
x = np.arange(len(studioss))
ax1 = plt.subplot(1, 1, 1)
w = 0.3
# plt.xticks(), will label the bars on x axis with the respective Studio names.
plt.xticks(x + w / 2, studioss["Studio"], rotation="vertical")
Count = ax1.bar(x, studioss["Movie_Count"], width=w, color="b", align="center")
# giving names to the axes and the title
ax1.set_ylabel("Movie Count")
ax1.set_xlabel("Studios")
plt.title("Number of movies per studio")
# To show the plot.
plt.show()
```

```python
# plot count and gross per studio
x = np.arange(len(studioss))
ax1 = plt.subplot(1, 1, 1)
w = 0.3
# plt.xticks(), will label the bars on x axis with the respective Studio names.
plt.xticks(x + w / 2, studioss["Studio"], rotation="vertical")
gross = ax1.bar(x, studioss["Total_Gross"], width=w, color="y", align="center")
# Set the X and Y axis label as gross.
ax1.set_xlabel("Studios")
ax1.set_ylabel("Total gross")
plt.title("Total gross per studio in millions")
# To show the plot.
plt.show()
```
